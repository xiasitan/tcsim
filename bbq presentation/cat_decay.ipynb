```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def diagonal_dm_elements(t):
    return np.exp(-t/2)
    
def off_diagonal_dm_elements(t):
    return np.exp(-t/16)
```

```python
t_list = np.linspace(0, 20, 100)
```

```python
plt.plot(t_list, diagonal_dm_elements(t_list), label="diagonal", color = "#E65F00", linewidth=4.0,)
plt.plot(t_list, off_diagonal_dm_elements(t_list), label="off-diagonal", color = "#48A078",linewidth=4.0)
plt.tight_layout()
plt.savefig("cat_decay.png", dpi=300)
```
