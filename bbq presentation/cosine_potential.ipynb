```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
freq = 2*np.pi*1

t = np.linspace(-0.5, 0.5, 1000)
```

```python
fig, ax = plt.subplots()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
ax.plot(t, -np.cos(freq*t), linewidth=7.0, c = "#48A078")
fig.frameon=False
fig.savefig('cosine_potential.png', dpi=300, bbox_inches='tight')
```

```python
fig, ax = plt.subplots()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
ax.plot(t, t**2, linewidth=7.0, c = "#E4718B")
fig.frameon=False
fig.savefig('square.png', dpi=300, bbox_inches='tight')
```
